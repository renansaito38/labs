# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
# escreva o código aqui
imagem_jp = Image.open('/content/drive/My Drive/Colab Notebooks/Computação Grafica/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')
imagem_ct = Image.open('/content/drive/My Drive/Colab Notebooks/Computação Grafica/claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem_bz = Image.open('/content/drive/My Drive/Colab Notebooks/Computação Grafica/bailey-zindel-NRQV-hBF10M-unsplash.jpg')

plt.figure(figsize=(20,4))

plt.subplot(131) #131 1 linha 3 colunas posição 1
plt.title('Imagem jp')
plt.axis('off')
plt.imshow(imagem_jp)

plt.subplot(132) #132 1 linha 3 colunas posição 2
plt.title('Imagem ct')
plt.axis('off')
plt.axis('off')
plt.imshow(imagem_ct)

plt.subplot(133) #133 1 linha 3 colunas posição 3
plt.title('Imagem bz')
plt.axis('off')
plt.imshow(imagem_bz)

plt.show()   


```

```
imagem_jp_arr = np.array(imagem_jp)
imagem_ct_arr = np.array(imagem_ct)
imagem_bz_arr = np.array(imagem_bz)
imagem_bz_arr.shape
```

```
red = imagem_jp_arr[:,:,0]
green = imagem_jp_arr[:,:,1]
blue = imagem_jp_arr[:,:,2]

plt.figure(figsize=(12,4))

plt.subplot(131) #131 1 linha 3 colunas posição 1
plt.hist(red.ravel(), color='red')

plt.subplot(132) #132 1 linha 3 colunas posição 2
plt.hist(green.ravel(), color='green')

plt.subplot(133) #133 1 linha 3 colunas posição 3
plt.hist(blue.ravel(), color='blue')
```

```
red = imagem_ct_arr[:,:,0]
green = imagem_ct_arr[:,:,1]
blue = imagem_ct_arr[:,:,2]

plt.figure(figsize=(12,4))

plt.subplot(131) #131 1 linha 3 colunas posição 1
plt.hist(red.ravel(), color='red')

plt.subplot(132) #132 1 linha 3 colunas posição 2
plt.hist(green.ravel(), color='green')

plt.subplot(133) #133 1 linha 3 colunas posição 3
plt.hist(blue.ravel(), color='blue')
```

```
red = imagem_bz_arr[:,:,0]
green = imagem_bz_arr[:,:,1]
blue = imagem_bz_arr[:,:,2]

plt.figure(figsize=(12,4))

plt.subplot(131) #131 1 linha 3 colunas posição 1
plt.hist(red.ravel(), color='red')

plt.subplot(132) #132 1 linha 3 colunas posição 2
plt.hist(green.ravel(), color='green')

plt.subplot(133) #133 1 linha 3 colunas posição 3
plt.hist(blue.ravel(), color='blue')
```
